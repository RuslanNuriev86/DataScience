```python
stations = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Moscow_contest/Map/stations.csv')
stations = stations.drop('Unnamed: 0', axis=1)
stations = stations.drop(1, axis=0).reset_index(drop=True)
stations
```

```python
#load predictions, create barplots and turn them to into json
anoh = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Moscow_contest/Map/Predictions/prediction_anoh.csv')
anoh = anoh.dropna(axis=0)
anoh['data_time'] = pd.to_datetime(anoh['data_time'], format='%Y-%m-%d %H:%M:%S').dt.hour
anoh = anoh.set_index('data_time')
anoh_bar = vincent.Bar(anoh, height=200, width=400)
anoh_bar.axis_titles(x ='hours', y ='concentration')
anoh_bar.legend(title='Indicators Day Forcast')
anoh_data = json.loads(anoh_bar.to_json())
anoh.head()
```

```python
#create the map
loc = 'Air quality monitoring map of Moscow'
title_html = '''
             <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc) 
map = folium.Map(location=[55.82, 37.525], tiles='openstreetmap', zoom_start=10)

mk_anoh = features.Marker([stations['latitude'][0], stations['longitude'][0]], tooltip = stations['eng_name'][0], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_kopt= features.Marker([stations['latitude'][2], stations['longitude'][2]], tooltip = stations['eng_name'][2], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_tur= features.Marker([stations['latitude'][7], stations['longitude'][7]], tooltip = stations['eng_name'][7], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_spir= features.Marker([stations['latitude'][6], stations['longitude'][6]], tooltip = stations['eng_name'][6], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_gleb= features.Marker([stations['latitude'][1], stations['longitude'][1]], tooltip = stations['eng_name'][1], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_ost= features.Marker([stations['latitude'][4], stations['longitude'][4]], tooltip = stations['eng_name'][4], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_prol= features.Marker([stations['latitude'][5], stations['longitude'][5]], tooltip = stations['eng_name'][5], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_mar= features.Marker([stations['latitude'][3], stations['longitude'][3]], tooltip = stations['eng_name'][3], icon=folium.Icon(color="darkblue", icon="cloud"))
mk_shab= features.Marker([stations['latitude'][8], stations['longitude'][8]], tooltip = stations['eng_name'][8], icon=folium.Icon(color="darkblue", icon="cloud"))

anoh_v = features.Vega(anoh_data, width="100%", height="100%")
kopt_v = features.Vega(kopt_data, width="100%", height="100%")
tur_v = features.Vega(tur_data, width="100%", height="100%")
spir_v = features.Vega(spir_data, width="100%", height="100%")
gleb_v = features.Vega(gleb_data, width="100%", height="100%")
ost_v = features.Vega(ost_data, width="100%", height="100%")
prol_v = features.Vega(prol_data, width="100%", height="100%")
mar_v = features.Vega(mar_data, width="100%", height="100%")
shab_v = features.Vega(shab_data, width="100%", height="100%")

anoh_p = folium.Popup("Hello")
kopt_p = folium.Popup("Hello")
tur_p = folium.Popup("Hello")
spir_p = folium.Popup("Hello")
gleb_p = folium.Popup("Hello")
ost_p = folium.Popup("Hello")
prol_p = folium.Popup("Hello")
mar_p = folium.Popup("Hello")
shab_p = folium.Popup("Hello")

mk_anoh.add_child(anoh_p)
mk_kopt.add_child(kopt_p)
mk_tur.add_child(tur_p)
mk_spir.add_child(spir_p)
mk_gleb.add_child(gleb_p)
mk_ost.add_child(ost_p)
mk_prol.add_child(prol_p)
mk_mar.add_child(mar_p)
mk_shab.add_child(shab_p)

anoh_p.add_child(anoh_v)
kopt_p.add_child(kopt_v)
tur_p.add_child(tur_v)
spir_p.add_child(spir_v)
gleb_p.add_child(gleb_v)
ost_p.add_child(ost_v)
prol_p.add_child(prol_v)
mar_p.add_child(mar_v)
shab_p.add_child(shab_v)

map.add_child(mk_anoh)
map.add_child(mk_kopt)
map.add_child(mk_tur)
map.add_child(mk_spir)
map.add_child(mk_gleb)
map.add_child(mk_ost)
map.add_child(mk_prol)
map.add_child(mk_mar)
map.add_child(mk_shab)

map.get_root().html.add_child(folium.Element(title_html))

map.save('forecast_map.html')
map
```
