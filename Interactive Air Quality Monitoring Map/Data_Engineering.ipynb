```python
#add profilemer data to datasets of stations
os.chdir("/content/drive/MyDrive/Colab Notebooks/Moscow_contest/")
datasets = [anoh, but, gleb, kopt, mar, ost, prol, spir, tur, shab]
datasets_name = ['anoh', 'but', 'gleb', 'kopt', 'mar', 'ost', 'prol', 'spir', 'tur', 'shab']
datasets_dict = dict(zip(datasets_name, datasets))

wind = wind.rename({'Дата и время': 'data_time'}, axis=1)
wind['data_time'] = wind['data_time'].astype(object)
wind = wind.set_index('data_time')

for name, data in datasets_dict.items():
  data = data.rename({'Дата и время': 'data_time'}, axis=1)
  data['data_time'] = data['data_time'].astype(object)
  data = data.merge(prof, on='data_time', how='left')
  data = data.merge(wind, on='data_time', how='left')
  print(data.shape)
  #save to files
  data.to_csv('{}.csv'. format(name))
```
