```python
#main function 
def main():
  updater = Updater(keys.API_KEY, use_context=True)
  dp = updater.dispatcher

  #dp.add_handler(CommandHandler('start'), start_command())
  dp.add_handler(MessageHandler(Filters.text, handle_message))
  updater.start_polling()
  updater.idle()

main()
```
